# Mel filterbank log spectrum

```python
Mspec = logMelSpectrum(FFT, example['samplingrate'])
plt.pcolormesh(Mspec)
plt.show()
```

# Cosine Transofrm and Liftering

```python
Mfcc = cepstrum(Mspec, 13)
plt.pcolormesh(Mfcc)
plt.show()
```

```python
#check
plt.pcolormesh(example['mfcc'])
plt.show()
```

```python
Lmfcc = lifter(Mfcc)
plt.pcolormesh(Lmfcc)
plt.show()
```

```python
#check
plt.pcolormesh(example['lmfcc'])
plt.show()
```

# Feature Correlation

```python
mfcc_utterances = []
mspec_utterances = []
for samples in data:
    mfcc_utterances.append(mfcc(samples["samples"]))
    mspec_utterances.append(mspec(samples["samples"]))

    
_mfcc = np.vstack(mfcc_utterances).T
_mspec = np.vstack(mspec_utterances).T

#print(len(_mfcc.T))

corr_mfcc = np.corrcoef(_mfcc)
corr_mspec = np.corrcoef(_mspec)

plt.pcolormesh(corr_mfcc)
plt.show()
plt.pcolormesh(corr_mspec)
plt.show()
```
